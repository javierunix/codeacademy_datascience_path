```python
import pandas as pd 
from matplotlib import pyplot as plt 
import numpy as np
```

```python
coaster_ranking_wood = pd.read_csv('golden_ticket_award_winner_wood.csv')
coaster_ranking_wood.head(20)
```

```python
coaster_ranking_steel = pd.read_csv('golden_ticket_award_winner_steel.csv')
coaster_ranking_steel.head(20)
```

```python
# Compare Rankings Of 'Boulder Dash' and 'El Toro'
```

```python
# Select 'Boulder Dash'
boulder_df = coaster_ranking_wood[coaster_ranking_wood['Name'] == 'Boulder Dash']
boulder_df
```

```python
# Select 'El Toro'
toro_df = coaster_ranking_wood[coaster_ranking_wood['Name'] == 'El Toro']
toro_df
```

```python
# There are two roller coaster called 'El Toro'. We are going to select the Six Flags Great Adventure's Park
toro_df_new = toro_df[toro_df['Park'] == 'Six Flags Great Adventure']
toro_df_new
```

```python
# Time Series Plot of 'Boulder Dash' and 'El Toro'

ax = plt.subplot()

plt.plot(toro_df_new['Year of Rank'], toro_df_new['Rank'])
plt.plot(boulder_df['Year of Rank'], boulder_df['Rank'], color='red')

ax.invert_yaxis()  # labels read top-to-bottom
plt.legend(['El Toro', 'Boulder Dash'])
plt.ylabel('Ranking')
plt.xlabel('Year')
plt.title('Top Rated Roller Coasters')

plt.show()
```
